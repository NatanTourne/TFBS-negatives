```python
import pandas as pd
```

```python
A549 = "/data/home/natant/Negatives/Data/maxATAC/ATAC_Peaks_hg38/A549_ext40_peaks.narrowPeak"
```

```python
A549_df = pd.read_csv(A549, sep="\t", header=None)
A549_df.columns = ["chr", "start", "stop", "name", "score", "strand", "signal value", "p value", "q value", "peak annotation"]
```

```python
A549_df["delta"] = A549_df["stop"] - A549_df["start"]
```

```python
A549_df["delta"].mean()
```

```python
import matplotlib.pyplot as plt
plt.hist(A549_df["delta"], bins=50)
plt.xlabel('Peak Length')
plt.ylabel('Frequency')
plt.title('Histogram of Peak Length')
plt.show()
```
