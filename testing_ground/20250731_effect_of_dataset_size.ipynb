```python
import pickle
import numpy as np

# Load the pickle file
with open('/data/home/natant/Negatives/TFBS_negatives/testing_ground/20250613_NUMBER_OF_POSITIVES_PER_GROUP.pkl', 'rb') as f:
    data = pickle.load(f)

print(data)
```

```python
import pandas as pd

# Create a dictionary to store dataframes for each cell line
dataframes = {}

# Iterate through each cell line
for cell_line, tf_data in data.items():
    # Create a list to store the data for this cell line
    rows = []
    
    # Iterate through each TF and extract the three values
    for tf_name, values in tf_data.items():
        rows.append({
            'TF': tf_name,
            'Split_1': values[0],
            'Split_2': values[1],
            'Split_3': values[2]
        })
    
    # Create dataframe for this cell line
    df = pd.DataFrame(rows)
    df.set_index('TF', inplace=True)
    
    # Store in dictionary
    dataframes[cell_line] = df
```
