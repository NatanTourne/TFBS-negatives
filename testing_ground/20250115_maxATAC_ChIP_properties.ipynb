```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
folder_path = "/data/home/natant/Negatives/Data/maxATAC/ChIP_Binding_File"
bed_files = glob.glob(folder_path + "/*.bed")

mean_delta = []
for bed_path in bed_files:
    chip_data = pd.read_csv(bed_path, sep = " ", header=None)
    chip_data.columns = ["chr", "start", "end", "TF"]
    chip_data["delta"] = chip_data["end"] - chip_data["start"]
    mean_delta.append(chip_data["delta"].mean())
```

```python
plt.hist(mean_delta, bins=50)
plt.xlabel('Mean Delta')
plt.ylabel('Frequency')
plt.title('Histogram of Mean Delta')
plt.show()
```
