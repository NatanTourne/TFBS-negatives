```python
import py2bit
genome_loc = "/data/home/natant/plmBind/Data/hg38.2bit"
```

```python
tb = py2bit.open(genome_loc)
```

```python
chromosomes = tb.chroms()
chromosomes
```

```python
chrom_names = [chrom for chrom in chromosomes.keys() if 'chrUn' not in chrom and 'chrM' not in chrom]
chrom_names
```

```python
from collections import defaultdict

# Initialize a dictionary to store the total lengths
chrom_total_lengths = defaultdict(int)

# Iterate through the chromosomes dictionary
for chrom, length in chromosomes.items():
    # Extract the main chromosome name (e.g., 'chr1' from 'chr1_KI270706v1_random')
    main_chrom = chrom.split('_')[0]
    # Add the length to the corresponding main chromosome
    chrom_total_lengths[main_chrom] += length

# Convert the defaultdict to a regular dictionary
chrom_total_lengths = dict(chrom_total_lengths)
```

```python
# Remove 'chrUn' and 'chrM' from the dictionary
chrom_total_lengths.pop('chrUn', None)
chrom_total_lengths.pop('chrM', None)

chrom_total_lengths
```

```python
import random

# Shuffle the keys of chrom_total_lengths
shuffled_keys = list(chrom_total_lengths.keys())
random.shuffle(shuffled_keys)
shuffled_keys
part1 = shuffled_keys[:len(shuffled_keys)//3]
part2 = shuffled_keys[len(shuffled_keys)//3:2*len(shuffled_keys)//3]
part3 = shuffled_keys[2*len(shuffled_keys)//3:]
```

```python
part1 = shuffled_keys[:len(shuffled_keys)//3]
part2 = shuffled_keys[len(shuffled_keys)//3:2*len(shuffled_keys)//3]
part3 = shuffled_keys[2*len(shuffled_keys)//3:]
```

```python
# Calculate total lengths for each part
part1_total_length = sum(chrom_total_lengths[chrom] for chrom in part1 if chrom in chrom_total_lengths)
part2_total_length = sum(chrom_total_lengths[chrom] for chrom in part2 if chrom in chrom_total_lengths)
part3_total_length = sum(chrom_total_lengths[chrom] for chrom in part3 if chrom in chrom_total_lengths)

part1_total_length, part2_total_length, part3_total_length
```

```python
part3
```

```python
# Create new lists to include alts for each group
part1_with_alts = part1[:]
part2_with_alts = part2[:]
part3_with_alts = part3[:]

# Iterate through the chromosomes dictionary
for chrom in chromosomes.keys():
    main_chrom = chrom.split('_')[0]  # Extract the main chromosome name
    if main_chrom in part1:
        part1_with_alts.append(chrom)
    elif main_chrom in part2:
        part2_with_alts.append(chrom)
    elif main_chrom in part3:
        part3_with_alts.append(chrom)

# Remove duplicates and sort the lists
part1_with_alts = sorted(set(part1_with_alts))
part2_with_alts = sorted(set(part2_with_alts))
part3_with_alts = sorted(set(part3_with_alts))

part1_with_alts, part2_with_alts, part3_with_alts
```
